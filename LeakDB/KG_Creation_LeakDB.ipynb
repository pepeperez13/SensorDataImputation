```python
from rdflib import Graph
from wdn_knowledge_graph.knowledge_graph import create_knowledge_graph_from_inp, networkx

from pathlib import Path

ontology_file = 'wdn_ontology.ttl'

# .inp file paths for all scenarios
inp_paths = [
    fr"D:\Master\Thesis\Code\LeakDB\LeakDB-master\CCWI-WDSA2018\Benchmarks\Hanoi_CMH\Scenario-{i}\Hanoi_CMH_Scenario-{i}.inp"
    for i in range(1, 11)
]

# Example: print all .inp file paths
for i in range(1,11):
    # Establish specific file names
    output_file = f'knowledge_graph{i}-LeakDB.ttl'
    inp_file = inp_paths[i-1]

    # Create knowledge graph
    knowledge_graph = create_knowledge_graph_from_inp(inp_file, destination=output_file)

    # Convert the RDF knowledge graph to a NetworkX graph for further processing
    networkx_graph = networkx(knowledge_graph)
    
    print(networkx_graph)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Assume G is your graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(networkx_graph, seed=42)  # You can also try nx.kamada_kawai_layout(G) for better spacing

# Draw the graph with labels
nx.draw(networkx_graph, pos, with_labels=True, node_size=500, font_size=8, arrows=True)

plt.title("Simple Graph Visualization")
plt.axis('off')
plt.show()
```
